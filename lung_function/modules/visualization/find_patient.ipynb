```python
import numpy as np 
import pandas as pd
```

```python
mode = 'valid'
id_1out_ls = [2543, 2541, 2539, 2537, 2535]
id_4out = 2751
id_ls = id_1out_ls + [id_4out]

df_err_ls = []
for id in id_ls:
    pred_fpath = f"/home/jjia/data/lung_function/lung_function/scripts/results/experiments/{id}/{mode}_pred.csv"
    label_fpath = f"/home/jjia/data/lung_function/lung_function/scripts/results/experiments/{id}/{mode}_label.csv"

    df_pred = pd.read_csv(pred_fpath,index_col=0)
    df_label = pd.read_csv(label_fpath,index_col=0)

    df_err = (df_pred - df_label).abs()  # absolute error
    if len(df_err.columns) > 1:
        for i in df_err.columns:
            df_new = df_err[i].to_frame()
            # print(df_err[i])
            df_err_ls.append(df_new)
    else:
        df_err_ls.append(df_err)
```

```python
for err in df_err_ls:
    err['order'] = err.rank().astype(int)

```
