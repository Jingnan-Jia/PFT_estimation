```python
import matplotlib.pyplot as plt

def draw_nested_pie(ax, values, labels, colors):
    width = 1 / (len(values) + 1)
    for i, (value, label, color) in enumerate(zip(values, labels, colors)):
        radius = 1 - (i * width)
        draw_single_pie(ax, value, label, color, radius, width)

def draw_single_pie(ax, value, label, color, radius, width):
    percentage = value
    start_angle = 90

    ax.pie([percentage, 1 - percentage],
           radius=radius,
           startangle=start_angle,
           colors=[color, 'none'],
           wedgeprops={'width': width})
    ax.text(0,
            radius - (width / 2),
            f'{label}: {percentage * 100:.2f}%',
            ha='center',
            va='center',
            fontsize=10,
            rotation=start_angle - 90 - (percentage * 360 / 2))

# 数据和标签
data = {
    'DLCOc': [0.71, 0.59, 0.5, 0.45, 0.51, 0.57],
    'FEV1': [0.77, 0.62, 0.58, 0.59, 0.53, 0.55],
    'FVC': [0.83, 0.73, 0.69, 0.69, 0.62, 0.64],
    'TLC': [0.82, 0.78, 0.68, 0.69, 0.66, 0.68]
}
labels = ['Task 1', 'Task 2', 'Task 3', 'Task 4', 'Task 5', 'Task 6']
colors = ['darkgreen', 'green', 'limegreen', 'lightgreen', 'lime', 'greenyellow']

# 创建画布
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
fig.subplots_adjust(wspace=0.1, hspace=0.1)

# 绘制环形进度条
for i, (title, values) in enumerate(data.items()):
    row, col = divmod(i, 2)
    ax = axs[row, col]
    draw_nested_pie(ax, values, labels[:len(values)], colors)
    ax.set_title(title, fontsize=14)
    ax.axis('equal')
    ax.set_axis_off()

# 设置大标题
fig.suptitle('Nested Circular Progress Bars with 4 Tasks', fontsize=20)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def draw_single_pie(ax, value, label, color, radius, width):
    percentage = value
    start_angle = 90

    ax.pie([percentage, 1 - percentage],
           radius=radius,
           startangle=start_angle,
           colors=[color, 'none'],
           wedgeprops={'width': width})
    
    end_angle = start_angle - percentage * 360
    x = (radius - width / 2) * np.cos(np.radians(end_angle))
    y = (radius - width / 2) * np.sin(np.radians(end_angle))
    
    ax.text(x, y,
            f'{value}',
            ha='center',
            va='center',
            fontsize=10)

# 创建画布
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# 调整子图之间的间距
fig.subplots_adjust(wspace=0.1, hspace=0.1)

# 绘制环形进度条
for i, (title, values) in enumerate(data.items()):
    row, col = divmod(i, 2)
    ax = axs[row, col]
    draw_nested_pie(ax, values, labels[:len(values)], colors)
    ax.set_title(title, fontsize=14)
    ax.axis('equal')
    ax.set_axis_off()

# 设置大标题
fig.suptitle('Nested Circular Progress Bars with 4 Tasks', fontsize=20)

# 创建图例
legend_labels = ['Whole CT', 'Both lungs', 'Left lung', 'Right lung', 'Vessel', 'Binary vessel']
legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, legend_labels)]
fig.legend(handles=legend_patches, loc='upper center', ncol=len(legend_labels), fontsize=12, bbox_to_anchor=(0.5, 0.95))

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
paths = [path1, path2, path3, path4, path5, path6]
dataframes = [(pd.read_csv(path)-df_label).abs() for path in paths]

# 计算均值和标准差
means = {}
stds = {}
for column in ['DLCOc_SB', 'FEV1', 'FVC', 'TLC_He']:
    means[column] = [df[column].mean() for df in dataframes]
    stds[column] = [df[column].std() for df in dataframes]

# 绘制带有误差线的条形图
fig, axs = plt.subplots(1, 4, figsize=(18, 6), sharey=False)
width = 0.3

for idx, (ax, (column, mean, std)) in enumerate(zip(axs, zip(means.keys(), means.values(), stds.values()))):
    x = np.arange(len(paths))
    ax.bar(x, mean, width, label=column, yerr=std, capsize=5, color='lightblue')
    
    # 设置x轴标签
    ax.set_xticks(x)
    legend_labels = ['Whole CT', 'Both lungs', 'Left lung', 'Right lung', 'Vessel', 'Binary vessel']

    ax.set_xticklabels(legend_labels, rotation=30)
    # 设置标题和轴标签
    if column=='DLCOc_SB':
        column = 'DLCOc'
    if column=='TLC_He':
        column = 'TLC'
    ax.set_title(column)
    if idx == 0:
        ax.set_ylabel('Mean Values for DLCOc')
    elif idx == 1:
        ax.set_ylabel('Mean Values for FEV1, FVC, and TLC')

# 设置大标题
fig.suptitle('Mean Values with Standard Deviation for DLCOc, FEV1, FVC, and TLC from 6 CSV Files', fontsize=16)

# 显示图形
plt.show()
```
